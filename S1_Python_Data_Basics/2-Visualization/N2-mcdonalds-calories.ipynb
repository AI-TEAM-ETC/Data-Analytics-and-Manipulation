```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
import seaborn as sns
```

## Read Data

```python
#If running on Kaggle
# path = "../input/nutrition-facts/menu.csv"

#If Running from repo
path = "./data/menu.csv"

df_original = pd.read_csv(path)
```

```python
df = df_original.copy()
```

```python
df.head()
```

```python
df.columns
```

```python
columns = ['Category', 'Item', 'Serving Size', 'Calories',
       'Total Fat', 'Cholesterol',
       'Carbohydrates', 'Dietary Fiber','Sugars', 'Protein',]

df = df[columns]
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Category'].value_counts()
```

## Visualization

### All Item Categories

#### Categorical Columns

```python
sns.countplot(data=df, y='Category').set_title('Count of items in each category')
```

### Numeric Columns

```python
sns.displot(data=df, x='Calories')
```

```python
sns.displot(data=df, x='Calories', hue='Category', kind='kde')
```

```python
sns.boxplot(data=df, x='Calories')
```

```python
sns.boxplot(data=df, y='Category', x='Calories')
```

```python
sns.scatterplot(data=df, x='Calories', y='Protein')
```

```python
sns.scatterplot(data=df, x='Calories', y='Protein', hue='Category')
```
