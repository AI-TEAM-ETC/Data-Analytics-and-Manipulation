# Filtering Data with Pandas

## Import Libraries

```python
import pandas as pd
import numpy as np
```

```python
pd.__version__
```

### Read CSV File to Dataframe

```python
#If running on Kaggle
# path = "../input/titanic/train.csv"

#If Running anywhere else
path = "https://github.com/datasciencedojo/datasets/raw/master/titanic.csv"

df = pd.read_csv(path)
```

```python
df.head()
```

### Column Manipulation

```python
df['PassengerId'] = 0
```

```python
df['Cabin'] = 'AAA'
```

```python
df['Sex'] = df['Sex'].replace(['male', 'female'], ['m', 'f'])
```

```python
df['Sex'] = df['Sex'].replace({
    'male': 'm',
    'female': 'f',
})
```

```python
df.head()
```

## Groupby 
#### Very Important!!!

#### Groupby Count

```python
df.groupby('Pclass').count()
```

```python
df.groupby('Pclass')['Name'].count()
```

```python
df[['Pclass']].value_counts()
```

#### Groupby Mean or Median

```python
df.groupby('Pclass').mean()
```

```python
df.groupby('Pclass')[['Age']].mean()
```

```python
df.groupby('Pclass')[['Age']].median()
```

```python
df.groupby('Pclass')[['Age', 'Fare']].mean()
```

#### Groupby Min & Max

```python
df[['Pclass', 'Age']].groupby('Pclass').min('Age')
```

```python
df[['Pclass', 'Age']].groupby('Pclass').max('Age')
```

#### Other Aggregation Functions

```python
df[['Pclass']].value_counts()
```

```python
df[['Pclass', 'Fare']].groupby('Pclass').agg(np.sum)
```

#### Multiple Aggregations

```python
df[['Pclass', 'Age']].groupby('Pclass').agg(['min', 'max', 'mean', 'median', np.sum])
```

#### Multiple Columns

```python
df.groupby('Pclass')[['Age', 'Fare']].median()
```

```python
df[['Pclass', 'Age', 'Fare']].groupby('Pclass').agg(['mean', 'median'])
```

#### Other Way To Use

```python
df.groupby('Sex')['Age'].agg(['mean', 'median', 'min', 'max'])
```

```python
df[['Pclass', 'Age']].groupby('Pclass').boxplot()
```

### Multiple Groupby

```python
df[['Name','Pclass']].groupby('Pclass').count()
```

```python
df.groupby(['Pclass', 'Sex'])[['Name']].count()
```

```python
df.groupby(['Sex', 'Pclass'])[['Name']].count()
```

```python
df.groupby(['Sex', 'Pclass'])[['Age']].median()
```

```python
df.groupby(['Pclass', 'Sex'])[['Age']].median()
```

```python
df.groupby(['Pclass', 'Sex'])[['Name']].count().reset_index()
```

### Sorting Values

```python
df.sort_values(by='Age')
```

```python
df.sort_values(by='Age', ascending=False)[0:10]
```

```python
df.sort_values(by=['Age', 'Name'], ascending=False)[0:10]
```

```python
df.sort_values(by=['Name', 'Age'], ascending=False)[0:10]
```

```python
df.sort_values(by=['Cabin', 'Age'])[0:10]
```

```python
df.sort_values(by=['Name'])
```
